```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_all(pathname, data, x):
    fig, axes = plt.subplots(1, 3, figsize=(14, 5))
    
    sns.set_theme(style="whitegrid")
    fig.suptitle(f"impact on the {pathname} dataset")
    #axes[0].set_title(pathname)
    sns.barplot(ax=axes[0], x=x, y='accuracy', data=data, capsize=.1)
    axes[0].set(ylim=(0, 1))
    #plt.show()
    
    #axes[1].set_title(pathname)
    sns.barplot(ax=axes[1], x=x, y='ari', data=data, capsize=.1)
    axes[1].set(ylim=(0, 1))
    #plt.show()
    
    #axes[2].set_title(pathname)
    sns.barplot(ax=axes[2], x=x, y='silhouette', data=data, capsize=.1)
    axes[2].set(ylim=(0, 1))
    #plt.show()
    fig.savefig(f"test/{x}/{pathname}.png")
```

```python
pathname = "kolodziejczyk"
norm = pd.read_csv(f'results/{pathname}_norm.csv')
plot_all(pathname, norm, "norm")
epochs = pd.read_csv(f'results/{pathname}_epochs.csv')
plot_all(pathname, epochs, "encode_epochs")
lr = pd.read_csv(f'results/{pathname}_lr.csv')
plot_all(pathname, lr, "lr")
wdecay = pd.read_csv(f'results/{pathname}_wdecay.csv')
plot_all(pathname, wdecay, "wdecay")
```
